```python
inputs = [1,2,3,4,5];
weights =[[0.2,0.8,-0.5,1.0],[0.5,-0.91,0.26,-0.5],[-0.26,0.27,0.17,0.87],[-0.26,0.27,0.17,0.87]];
biases =[2,3,4,0.5];


outputwaa=[];

for ek_neuron_ka_weigghts ,ek_neuron_ka_biases in zip(weights,biases):
  ek_neuron_ka_output=0;

  for n_input ,weight in zip(inputs, ek_neuron_ka_weigghts):
    ek_neuron_ka_output = ek_neuron_ka_output+ n_input * weight

    ek_neuron_ka_output =ek_neuron_ka_output+ ek_neuron_ka_biases;

  outputwaa.append(ek_neuron_ka_output);

  print(outputwaa);


```
