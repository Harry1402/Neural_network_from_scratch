```python
import numpy as np ;

X=[[2,3,4,5],
   [3,4,5,6],
   [2,3,4,5]]; # (3,4)

class Layer_Dense:
  def __init__(self,n_inputswa,n_neuronswa):
    self.weights =0.10 * np.random.randn(n_inputswa ,n_neuronswa) #(rows,colums)
    self.biases=np.zeros((1,n_neuronswa))
  def forward(self ,inputs):
    self.output=np.dot(inputs,self.weights)+self.biases

layer1 = Layer_Dense(4,5)
layer2 = Layer_Dense(5,2)
layer3 =Layer_Dense(2,5)

layer1.forward(X)
layer2.forward(layer1.output) # Call the forward method to compute the output
layer3.forward(layer2.output)
print(layer3.output);
```
